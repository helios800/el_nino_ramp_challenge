# Introduction <a class="anchor" id="introduction"></a>


The El Nino phenomenon is a sea surface temperature (SST) anomaly in the Pacific Ocean. It is part of the
of the El Nino Southern Oscillation (ENSO) phenomenon. El Nino is a pseudo-oscillatory phenomenon, shifting from a cold phase to a warm phase.
This causes large-scale air movements in the tropics, with global effects. This
has a direct impact on the economy and the weather, affecting fish migrations and therefore the economy of the concerned countries.
The change between the two phases is periodic, with irregular periods ranging from 2 to 7 years. This phenomenon 
reaches its peak around December. El Nino's influence is global. For example, it prevents the upwelling of cold water on the west coast of Peru, as well as rainfall deficits in Australia and Indonesia, leading to fires and droughts. In addition, global temperatures seem to be abnormally high after a strong El Nino
El Nino phenomenon, as it was the case in 1998. 





<img src="https://meteo45.com/images/el-nino.png" alt="Alternative text" width="600px" />




The warming phase of the sea surface temperature is known as El Niño and the cooling phase as La Niña. The Southern Oscillation is the accompanying atmospheric component, which is coupled with the sea temperature change. El Niño is associated with higher than normal air sea level pressure over Indonesia, Australia and across the Indian Ocean to the Atlantic. La Niña has roughly the reverse pattern: high pressure over the central and eastern Pacific and lower pressure through much of the rest of the tropics and subtropics. 


El Niño and La Niña affect the global climate and disrupt normal weather patterns, which as a result can lead to intense storms in some places and droughts in others. El Niño events cause short-term (approximately 1 year in length) spikes in global average surface temperature while La Niña events cause short term surface cooling.Therefore, the relative frequency of El Niño compared to La Niña events can affect global temperature trends on decadal timescales. Developing countries dependent upon agriculture and fishing, particularly those bordering the Pacific Ocean, are the most affected.

In climate change science, ENSO is known as one of the internal climate variability phenomena. Future trends in ENSO due to climate change are uncertain, although climate change exacerbates the effects of droughts and floods. The IPCC Sixth Assessment Report summarized the state of the art of research in 2021 into the future of ENSO as follows: "In the long term, it is very likely that the precipitation variance related to El Niño–Southern Oscillation will increase" and "It is very likely that rainfall variability related to changes in the strength and spatial extent of ENSO teleconnections will lead to significant changes at regional scale".



We will here therefore aim at the prediction of this time serie. It falls into the regression part of machine learning, and we will here try to implement different methods to do this, and compare it to a simple classical ARMA (autoregressive moving average) model, from classical time series analysis.
Having reliable predictions of this phenomenon may, as highlighted before, help predicting rare events like storms, and is of principal importance for fishermans for instance.

## Importation of the data

```python
url = "https://psl.noaa.gov/data/correlation/nina3.anom.data"
df = pd.read_csv(url,skiprows=1,sep='\s+')
df = df.iloc[:-3]

new_column_names = ['Year',
                    'January',
                    'February',
                    'March',
                    'April',
                    'May',
                    'June',
                    'July',
                    'August',
                    'September',
                    'October',
                    'November',
                    'December']

df.columns = new_column_names

cols_to_convert = df.columns[1:]
df[cols_to_convert] = df[cols_to_convert].astype(float)
```

```python
df
```

this dataset will not be easy to handle with this current form. Let's transform it in a 2 columns dataframe, with dates and values. Let's also remove the values that appears to be irrelevant, like the 1949 data.

```python
glob_array = []
for k in range(1,75):
    for j in range(1,13):
        glob_array.append(df.iloc[k,j])
data = np.array(glob_array)

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
arr = data
years = range(1950, 2024)

data3 = []

# Iterate over years and months to construct the data
for year in years:
    for month, value in zip(months, arr[:12]):
        data3.append([f"{year}-{months.index(month) + 1}", value])  # Construct the date format as 'YYYY-MM'
    arr = arr[12:]

# Create a DataFrame from the data
df3 = pd.DataFrame(data3, columns=['Date', 'SST Anomaly'])
df3['Date'] = pd.to_datetime(df3['Date'])
df3['Date'] = df3['Date'].dt.strftime('%Y-%m')

df3
```

# Data exploration <a class="anchor" id="exploration"></a>

We can now visualize the data:

```python
colors = ['red' if value < 0 else 'orange' for value in df3['SST Anomaly']]
plt.figure(figsize = (8,6))
ax = sns.barplot(data=df3, y='SST Anomaly', x='Date',palette=colors)


step = len(df3) // 10  
ax.set_xticks(range(0, len(df3), step))
ax.set_xticklabels(df3['Date'][::step], rotation=45) 


plt.grid()
plt.ylabel('SST Anomaly')
plt.title('SST anomaly since 1950')
plt.tight_layout()
plt.show()
```

As mentionned before, we can clearly see an oscillation between two El Nino/La NIna phase, respectively a positive and a negative phase. Let's now check if the data is standardized: 

```python
mean = df3['SST Anomaly'].mean()
std = df3['SST Anomaly'].std()

print('mean:',mean,'std:',std)
```

We can see here that the data is not perfectly standardize. Let's therefore do it:

```python
standardized = (df3['SST Anomaly'] - mean) / std

mean = standardized.mean()
std = standardized.std()

print('mean:',mean,'std:',std)
```

```python
df3['standardized anomaly'] = standardized
```

We therefore had to standardize the data. Let's however highlight that the data has already been detrended. Indeed, There is usually a clear (linear) trend in the sea surface temperature corresponding to global warming, which clearly do not appear here. We therefore have now a standard, detrended, deseasoned dataset. 

## autocorrelation


It's important to study the autocorrelation and the partial autocorrelation functions of our data before trying to predict it. Indeed, those functions will give us informations on possible patterns and also on the randomness of the time serie. Those two functions are mainly used to fit autoregressive models (AR) or moving average (MA) models (and obviously autoregressive moving average models (ARMA)) in classical time series analysis.

```python
acf = sm.tsa.acf(df3['standardized anomaly'], nlags=20)
pacf = sm.tsa.pacf(df3['standardized anomaly'], nlags=20)

# Plot ACF
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(range(len(acf)), acf)
plt.xlabel('Lag')
plt.ylabel('ACF')
plt.title('Autocorrelation Function (ACF)')

# Plot PACF
plt.subplot(1, 2, 2)
plt.bar(range(len(pacf)), pacf)
plt.xlabel('Lag')
plt.ylabel('PACF')
plt.title('Partial Autocorrelation Function (PACF)')

plt.tight_layout()
plt.show()
```

We can see here that there is a clear autocorrelation in our dataset. This is a good thing, it means that our data is not 'too random'. This is needed for statistical prediction models relying on the fact that the knowledge of past data can help predicting the future data. We can also highlight the inversion of the correlation at lag 10. 

## Stationnarity

Classical statistical models  rely on the stationnarity hypothesis. We will here try the Augmented Dickey Fuller Test (ADF). For more informations, check : https://en.wikipedia.org/wiki/Augmented_Dickey%E2%80%93Fuller_test. 

If the time serie is not stationnary, we can make it stationnary by differencing it, until it becomes stationnary.

```python
def check_stationarity(series):

    result = adfuller(series.values)

    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

    if (result[1] <= 0.05) & (result[4]['5%'] > result[0]):
        print("\u001b[32mStationary\u001b[0m")
    else:
        print("\x1b[31mNon-stationary\x1b[0m")
```

```python
check_stationarity(df3['standardized anomaly'])
```

Our data is already stationnary, which means that we can now try to predict it using regression models.

## train/test split

we will now split our data in a train/test fashion. Since we have limited data, we will try to keep enough training data to get accurate predictions. We therefore choose 700 training data, for 188 test data. Note that since it is a time serie regression problem, randomizing the dataset will not help us, since the prediction will rely only on the idea that the data is correlated in time.

```python
serie = df3['standardized anomaly'].values
```

```python
xtrain,xtest = serie[:700],serie[700:]
```

## Base model <a class="anchor" id="base_model"></a>

Since our serie is stationnary, we can first build a simple ARMA model to predict the future values of our sea surface temperature anomaly. Formally, considering our data as a realization of a random variable $X_t$, we can define an ARMA process as: 

\begin{align}
    \ X_t = \epsilon_t + \sum_i^p \phi_iX_{t-i} + \sum_k^q \theta_k\epsilon_{t-k}
\end{align}





An arma model gets two parameters:

-p: order of the autoregressive part, corresponding to the number of significant lags in the ACF plot <br>
-q: order of the moving average part, corresponding to the number of significant lags in the PACF plot.

According to our previous analysis, we can consider an ARMA(8,3) at first sight. 

```python
model_arma = ARIMA(xtrain, order=(8,3,0))
model_fit = model_arma.fit()
print(model_fit.summary())
```

Let's now compute the prediction error for a one time step ahead prediction on our test data. Like many regression models, we will use the RMSE as error metrics. 

```python
history = [x for x in xtrain]
predictions = list()


for t in range(len(xtest)):
     model_arma = ARIMA(history, order=(8,3,0))
     model_fit = model_arma.fit()
     output = model_fit.forecast()
     yhat = output[0]
     predictions.append(yhat)
     obs = xtest[t]
     history.append(obs)

        
        
rmse = np.sqrt(mean_squared_error(xtest, predictions))
print('Test RMSE: %.3f' % rmse)

plt.figure(figsize = (6,4))
plt.plot(xtest,color = 'red', label = 'test data')
plt.plot(predictions, color = 'black', linestyle = '--', label = 'one step ahead prediction')
plt.xlabel('time step')
plt.ylabel('index')

plt.title('one time step ahead prediction on test data')
plt.legend()
plt.show()
```

## base score on test data: RMSE = 0.312

We have therefore a RMSE skill of 0.312. We may say that we can correctly predict the index value of the following month, using the 8 previous one. Let's now check the proposed model. The proposed model should beat the ARMA model in the challenge. 

# Proposed model: LSTM <a class="anchor" id="new_model"></a>

We will here consider a classical machine learning model for time serie prediction: Long Short Term Memory (LSTM).
This model is based on Recurrent Neural Networks (RNN), and can deal with the vanishing gradient problem. It is therefore aimed at predicting values using longer input sequences. We hope that using 8 previous values to compute the following ones will be much better with LSTM than with ARMA, which is a linear model.


<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/6/63/Long_Short-Term_Memory.svg/470px-Long_Short-Term_Memory.svg.png" alt="Alternative text" width="600px" />




Firstly, we will have to create input data. To do so, we feed our input with 8 previous values, in order to predict the following one. We choose this number because of the autocorrelation, where a value is correlated up to the 9 following time steps.

```python
def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i+sequence_length])
        y.append(data[i+sequence_length])
    return np.array(X), np.array(y)
```

Let's construct our test and train data:

```python
sequence_length = 8  
X_train, y_train = create_sequences(xtrain, sequence_length)  #training data is now consisting of arrays of size 8
X_test, y_test = create_sequences(xtest, sequence_length) #same for test data

# Reshape data for LSTM (samples, time steps, features)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1]))  
```

```python
print('shape X_train:',np.shape(X_train))
```

this will therefore be the shape used for our test and training data for the ramp challenge, in the ramp-test. 

We can now build our simple LSTM model:

```python
model = Sequential()
model.add(LSTM(units=100, input_shape=(X_train.shape[1], 1)))
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=30, batch_size=50)
loss = model.evaluate(X_test, y_test)
print('Test Loss:', loss)
```

```python
plt.figure(figsize = (6,4))
plt.plot(model.predict(X_test).flatten(),color = 'black',linestyle = '--',label = 'one step prediction')
plt.plot(xtest[8:],color = 'red', label = 'true test data')
plt.grid()
plt.legend()
plt.xlabel('time steps')
plt.title('one time step ahead prediction on test data')
plt.ylabel('index')
plt.show()
```

same way than before, we check the RMSE between the true value and the one step ahead prediction.

```python
rmse = np.sqrt(mean_squared_error(xtest[8:], model.predict(X_test).flatten()))
print('Test RMSE: %.3f' % rmse)
```

we get a better RMSE skill than with the ARMA model for a one time step ahead prediction. Let's see how it goes for multiple time steps ahead. Note that due to the autocorrelation function, we will not be able to predict over 10 months ahead. 

# multiple time steps ahead prediction: Bonus analysis

Our challenge on RAMP will test only the performance on a one time step ahead prediction, but it is interesting to see the results on farther away in time predictions. 

```python
def multiple_step_prediction(starting_data, model, step_nb):
    current_data = np.expand_dims(starting_data, axis=0)
    
    for k in range(step_nb):
        predicted = model.predict(current_data).flatten()[0]
        print(predicted)
        current_data = list(current_data)
        del current_data[0]
        current_data.append(predicted)
        current_data =  np.expand_dims(np.array(current_data),axis = 0)
        
    final_step = current_data[-1]
    return final_step

def predict_multiple(test_data, model, step_nb): 
    prediction = []
    for k in range(len(test_data)): 
        prediction.append(multiple_step_prediction(test_data[k],model,step_nb))
        
    return prediction
```

let's build a 3 time steps ahead prediction:

```python
predicted = predict_multiple(X_test, model, 3)
```

```python
plt.plot(predicted, label = '3 time steps ahead prediction',color = 'black',linestyle = '--')
plt.plot(xtest[8:],label = 'true test data',color = 'red')
plt.xlabel('time steps')
plt.ylabel('index')
plt.legend()
plt.show()
```

We can see that our model fails to recreate the amplitude of our data. However, it seems to be able to reproduce the variations of the true test data. Then we can now lower a bit our expectations and try to predict the changes between the negative or positive phases, i.e the El Nino and La Nina events. 

# changes from positive to negative phase

As mentionned before, we are more interested in predicting the changes between positive or negative phase than the numerical value of the deviation. Let's now check if we can accurately predict the sign of the index. 

We will here therefore not only consider the RMSE as an error metric, but also the number of falsely classified events between El Nino ($sign(X_t) = 1$) and La Nina ($sign(X_t) = -1$). We will compute the evolution of this metric with respect to the number of months ahead we try to predict. 

```python
def class_error(true_data, prediction): 
    error  = 0 
    for k in range(len(prediction)): 
        if np.sign(prediction[k]) != np.sign(true_data[k]): 
            error += 1
    return error/len(prediction)
```

```python
classification_error = np.zeros((12))
rmse_error = np.zeros((12))

for ts in np.arange(1,12): 
    predicted = predict_multiple(X_test, model, ts)
    rmse_error[ts] = np.sqrt(mean_squared_error(xtest[8:], predicted))
    classification_error[ts] = class_error(xtest[8:],predicted)
```

```python
plt.figure(figsize = (6,4))
plt.scatter(np.arange(1,12),rmse_error[1:], color = 'red', marker = 'o', label = 'rmse error LSTM')
plt.plot(np.arange(1,12),rmse_error[1:], color = 'red', linestyle ='--')
plt.scatter(np.arange(1,12), classification_error[1:], color = 'black', marker = 'o', label = 'classification error LSTM')
plt.plot(np.arange(1,12),classification_error[1:], color = 'black', linestyle ='--')


plt.grid()
plt.legend()
plt.xlabel('month ahead predicted')
plt.ylabel('metrics score')
plt.show()
```

we can see here that eventhough the RMSE is increasing fast with the number of months ahead we want to predict, we can have a 'correct' prediction of the El Nino/El Nina phase up to 4 months, with 80% of accuracy. However, we fall to less than 50% of accuracy on this prediction when trying to predict more than 6 months, which means that the behaviour become random. 

# Submitting on RAMP: one time step ahead prediction <a class="anchor" id="submitting"></a>

As we said before, the performance tested for the challenge will only be on a 1 time step ahead prediction.
We will use the X_test, Y_test, X_train, Y_train as we introduced them in the LSTM part, using 8 months data to predict the index of the following month. Note that there is only one CV part, because it is not relevant for a time serie analysis. The submission should consist of a get_estimator() function: 

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.base import BaseEstimator

class LSTMEstimator(BaseEstimator):
    def __init__(self, units=100, optimizer='adam', loss='mean_squared_error', epochs=30, batch_size=50):
        self.units = units
        self.optimizer = optimizer
        self.loss = loss
        self.epochs = epochs
        self.batch_size = batch_size
        self.model = None

    def fit(self, X, y):
        self.model = Sequential()
        self.model.add(LSTM(units=self.units, input_shape=(X.shape[1], 1)))
        self.model.add(Dense(units=1))
        
        self.model.compile(optimizer=self.optimizer, loss=self.loss)
        self.model.fit(X, y, epochs=self.epochs, batch_size=self.batch_size)
        
        return self

    def predict(self, X):
        return np.reshape(self.model.predict(X),np.shape(self.model.predict(X))[0])
    
def get_estimator():
    return LSTMEstimator()
```
